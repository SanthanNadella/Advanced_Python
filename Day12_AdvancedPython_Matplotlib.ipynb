```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#sample data

x = [1,2,3]
y = [7,8,9]
z = [4,5,6]

plt.plot(x,y)
plt.show()
```

```python
#as further enhancement of the graph

x = [1,2,3]
y = [7,8,9]
z = [4,5,6]

plt.xlabel('sales')       #adding x label
plt.ylabel('revenue')     #adding y label
plt.title('Sales performance')   #adding title to graph
plt.plot(x,y)
plt.show()
```

```python
Adding a new plot layer x and z
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['This is x and y','This is x and z'])   #adding legend to show difference between the layer
plt.show()                                          #showing the graphic view
```
